```python
import matplotlib.pyplot as plt

# Datos proporcionados
cities = ["Los Angeles", "San Diego", "San Jose", "San Francisco", "Fresno"]
populations = [3884307, 1345895, 1000536, 837442, 509039]
colors = ["blue", "gray", "olive", "brown", "pink"]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(cities, populations, color=colors, alpha=0.7)

# Añadir texto con las poblaciones encima de las barras
for i, texto in enumerate(populations):
    plt.text(i, texto + 50000, f"{texto:,}", ha="center", fontsize=10)

# Añadir etiquetas y título
plt.title("Ciudades más pobladas de California")
plt.ylabel("Población en millones")
plt.show()
```

```python
# 2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

area_total = df['area_total_km2'].isnull().sum() # 5

df.dropna(subset=['area_total_km2'], inplace=True)

area_total = df['area_total_km2'].isnull().sum() # 0
```

```python
# Generar datos de ejemplo similares
area_total = df['area_total_km2']
# Crear el histograma
plt.figure(figsize=(10, 10))
plt.hist(area_total, bins=50, color="steelblue", alpha=0.7, label="Hist Area")

# Ajustar título, etiquetas y leyenda con formato específico
plt.title("Histograma del Área de las Ciudades", fontname="Bradley Hand ITC", fontsize=18)
plt.xlabel("Superficie ($km^2$)", fontsize=15, fontstyle="italic")
plt.ylabel("Número de Ciudades", fontsize=15, fontstyle="italic")
plt.legend(loc="upper right", fontsize=12)

# Ajustar formato de los ejes
plt.xticks(fontsize=18, rotation=0, fontname="Bradley Hand ITC")
plt.yticks(fontsize=18, fontname="Bradley Hand ITC")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# 3. El scatter de área de agua de un ciudad con su poblacion.
area_total_agua = df['area_water_km2'].isnull().sum() # 5

df.dropna(subset=['area_water_km2'], inplace=True)

area_total = df['area_water_km2'].isnull().sum() # 0

# Generar datos simulados
area_agua = df['area_water_km2']
poblacion = df['population_total']

# Crear el scatter plot
plt.figure(figsize=(10, 10))
plt.scatter(area_agua, poblacion, color="green", alpha=0.7)

# Añadir título y etiquetas
plt.title("Población por kilómetro cuadrado de agua")
plt.xlabel("Área de agua en la ciudad ($km^2$)")
plt.ylabel("Población (miles habitantes)")

# Anotar outliers
plt.annotate("Outliers", xy=(120, 1400000), xytext=(70, 3000000),
             arrowprops=dict(facecolor="black", arrowstyle="->"))
plt.annotate("", xy=(87, 3800000), xytext=(75, 3100000),
             arrowprops=dict(facecolor="black", arrowstyle="->"))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# 4. El boxplot de la altura de las ciudades.
elevacion = df['elevation_m']

# Crear el boxplot
plt.figure(figsize=(10, 10))
plt.boxplot(elevacion, vert=False, patch_artist=True, boxprops=dict(facecolor="red", color="black"))

# Añadir título y etiquetas
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Ciudades")
plt.ylabel("Altitud (m)")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
